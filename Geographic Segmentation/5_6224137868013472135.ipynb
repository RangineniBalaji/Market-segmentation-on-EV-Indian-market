```python
import pandas as pd
```

```python
df = pd.read_csv("EV Stats-1.csv")
df = df.drop(["Sl. No"],axis=1)
total = df.iloc[30]
df = df.drop([30,31,32,33,34])
print(total)
df
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

```python
states = list(df["State"])
values = list(df["Two Wheelers (Category L1 & L2 as per Central Motor Vehicles Rules"])

fig = plt.figure(figsize = (10, 5))
 
plt.bar(states, values,
        width = 0.4)
y_pos = range(len(states))
plt.xlabel("States")
plt.ylabel("No. of Vehicles")
plt.title("Category L1 & L2 as per Central Motor Vehicles Rules)")
plt.xticks(y_pos, states, rotation=90)
plt.show()
```
